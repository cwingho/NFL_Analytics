```python
pos = (23,9,7,9,15,23,12,6) 
ind = np.arange(len(pos))    
width = 0.5
  
fig = plt.subplots(figsize =(15, 5)) 
p1 = plt.bar(ind, pos, width) 

plt.ylabel('No. of injury') 
plt.xlabel('Position') 
plt.title('Number of injury per position') 
plt.xticks(ind, ('Linebacker','Defensive Lineman','Offensive Lineman','Running Back',
                 'Safety','Wide Receiver','Cornerback','Tight End')) 
plt.yticks(np.arange(0, 25, 2))
plt.tight_layout()
plt.savefig('../out/injury_n_per_pos.png')
# plt.show() 
```
